```python
keyword_array = ['coronaal', 'coronal','cor',  'long', 'lung', 'lungs', 'longen', 'Blanco', 'MRP','sagittal', 'sag', 'Scanogram', 'Schedel','skull', 'body' ]
```

```python
def remove_keywords(keyword_array, dataFrame, column = 'Series_description' ):
    '''
    Removes scan from dataframe if it contains a keyword
    is not case sensitive
    '''	
    
    binary = np.zeros(len(dataFrame))
    
    for keyword in keyword_array:
        index = dataFrame.index[dataFrame[column].str.contains(keyword, case=False) == True]
        binary[np.array(index)] = 1

    remove_index = np.where(binary == 1)[0]
    
    df_new = dataFrame.drop(remove_index)
    df_removed = dataFrame.iloc[remove_index]
    return df_new, df_removed
```

```python
df_new,df_removed = remove_keywords(keyword_array, dataFrame)

df_removed
```
